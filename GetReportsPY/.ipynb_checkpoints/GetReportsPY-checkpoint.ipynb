```python
import json
```

```python
def write_json(data, filename='new_reports.json'): 
    with open(filename,'w') as f: 
        json.dump(data, f, indent=4) 
```

```python
array = []

with open('reports.json') as f:
  data = json.load(f)
    
  for user in data:
    
    userReports = data[user]
    
    for reportID in userReports:  
        
        rep = userReports[reportID]
        print(rep)
        array.append(rep)
        print("\n")
        
print(array)
        
write_json(array)
```
